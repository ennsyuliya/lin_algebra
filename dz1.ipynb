__1.__ Исследуйте на линейную зависимость:

$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

Выразим одну функцию, через другую:

$$f_{4}(x)=f_{3}(x)-f_{2}(x)-f_{1}(x)=x+1-1-e^{x}=x-e^{x}.$$

Таким образом, можно сделать вывод, что функции имеют линейную зависимость

$$f_{4}(x)=f_{3}(x)+f_{2}(x)+f_{1}(x)=x^{2}+2x+1=(x+1)^{2}$$

Таким образом, можно сделать вывод, что функции имеют линейную зависимость

__1.__ Найдите скалярное произведение векторов $x, y \in \mathbb{R}$:<br>
а) $x=(0,-3, 6),~y=(-4, 7, 9);$<br>
б) $x=(7, -4, 0, 1),~y=(-3, 1, 11, 2).$

```python

import numpy as np
from numpy.linalg import norm
```

```python
a1 = np.array([0, -3, 6])
b1 = np.array([-4, 7, 9])
a2 = np.array([7, -4, 0, 1])
b2 = np.array([-3, 1, 11, 2])

print(f'Скалярное произведение a1 и b1:\n{a1.dot(b1)}')
print(f'Скалярное произведение a2 и b2:\n{a2 @ b2}')
```

__2.__ Найдите нормы векторов $(4, 2, 4)$ и $(12, 3, 4)$ и угол между ними.

```python
a = np.array([4, 2, 4])
b = np.array([12, 3, 4])

print(f'норма a {norm(a, ord=1)}')
print(f'норма b {norm(b, ord=1)}')
print(f' Норма Евклида {norm(a)}')
print(f' Норма Евклида {norm(b)}')

cos_phi = np.dot(a, b) / norm(a) / norm(b)
print(f'Косинус угла между векторами {cos_phi:.2f}')
print(f'Угол {np.arccos(cos_phi):.2f}')
```

в) $(1/2, -1/2, 0), (0, 1/2, 1/2), (0,0,1);$<br>
г) $(1,0,0),(0,1,0),(0,0,1)?$ 

```python
# а)  (1,0,0),(0,0,1);
a = np.array([1, 0, 0])
b = np.array([0, 0, 1])
print('Ортонормированный базис', np.dot(a, b))
```

```python
#б)  (1/2⎯⎯√,−1/2⎯⎯√,0),(1/2⎯⎯√,1/2⎯⎯√,0),(0,0,1); 
c = np.array([(1/(2**0.5)), (-1/(2**0.5)), 0])
d = np.array([(1/(2**0.5)), (1/(2**0.5)), 0])
b = np.array([0, 0, 1])
print('Ортонормированный базис')
print(np.dot(c, d))
print(np.dot(c, b))
print(np.dot(b, d))
print(f'{norm(c):.1f}, {norm(d):.1f}, {norm(b):.1f}')
```

```python
#в)  (1/2,−1/2,0),(0,1/2,1/2),(0,0,1); 
e = np.array([0.5, -0.5, 0])
f = np.array([0, 0.5, 0.5])
b = np.array([0, 0, 1])
print('Не ортогональный, не ортонормированный базис')
print(np.dot(e, f))
print(np.dot(e, b))
print(np.dot(b, f))
print(norm(e), norm(f), norm(b))
```
